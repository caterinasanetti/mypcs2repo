```python
from Bio import SeqIO
seqs = []
with open ('rosalind_long.txt','r') as fa:
    for seq_record  in SeqIO.parse(fa,'fasta'):
        seqs.append(str(seq_record.seq))
#overlap 
seq1=seqs[0]
k=int(1/2*(len(seq1))+1)
#loop suffix-prefix
for i in range(999999):
    newseq=seq1
    for j in range(k,2*(k-1)):
        for seq in seqs[1:]:
            if seq1[-j:] == seq[:j]:
                seq1 = seq1 + seq[j:]
    if newseq == seq1:
        break
#loop prefix-suffix
for i in range(999999):
    newseq=seq1
    for j in range(k,2*(k-1)):
        for seq in seqs[1:]:
            if seq[-j:]== seq1[:j]:
                seq1 = seq+ seq1[j:]
    if newseq == seq1:
        break
print(seq1)



```
