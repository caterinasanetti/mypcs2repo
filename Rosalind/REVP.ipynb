```python
from Bio import SeqIO
seq_name, seq_string = [], []
with open ('rosalind_revp.txt','r') as fa:
    for seq_record  in SeqIO.parse(fa,'fasta'):
        seq_string.append(str(seq_record.seq))
z=seq_string[0]
s="".join(z)
b=[]
for i in s:
    if i=="A":
        b.append("T")
    if i=="T":
        b.append("A")
    if i=="C":
        b.append("G")
    if i=="G":
        b.append("C")
r="".join(b)
n=4
for i in range(len(s)):
        while n>= 4 and n<=12 and i+n<=len(s):
            if (s[i:i+n]==(r[i:i+n])[::-1]):
                print(i+1,n)
            n += 1
        n=4


```
