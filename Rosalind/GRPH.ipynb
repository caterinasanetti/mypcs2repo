```python
from Bio import SeqIO
my_seqlist = []
ids=[]
with open ('rosalind_grph.txt','r') as fa:
    for seq_record  in SeqIO.parse(fa,'fasta'):
        my_seqlist.append(str(seq_record.seq))
        ids.append(seq_record.id)
a=[]
b=[]
for x in range(len(my_seqlist)):
    for y in range(len(my_seqlist)):
        if my_seqlist[x][-3:] == my_seqlist[y][:3]:
            if x != y:
                a.append(ids[x])
                b.append(ids[y])
for i in range(len(a)):
    print(a[i] + " " + b[i])

        
```
